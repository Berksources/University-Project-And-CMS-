```python
import requests
import json
import pandas as pd
from statsmodels.tsa.ar_model import AutoReg
import urllib3
import time
import schedule

urllib3.disable_warnings()


def predictTempature():
    UrltoGetTem = "http://localhost:49743/api/Data/TempatureData"

    weather = pd.read_json(UrltoGetTem)

    X = weather.dataValue

    train, test = X[1:len(X) - 1], X[len(X) - 1:]

    model = AutoReg(train, lags=29)

    model_fit = model.fit()

    predictions = model_fit.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

    vara = predictions.to_json()

    vara = json.loads(vara)

    valueOfPrediction = vara["999"]

    p = {"newDatas": valueOfPrediction}

    UrlToPostTemp = "http://localhost:49743/api/Result/TempatureResult"

    headers = {'content-type': 'application/json'}

    response = requests.post(UrlToPostTemp, headers=headers, data=json.dumps(p), verify=False)

    print(response, "Tempature")


def predictHumidity():
    UrltoGetTem = "http://localhost:49743/api/Data/HumidityData"

    weather = pd.read_json(UrltoGetTem)

    X = weather.dataValue

    train, test = X[1:len(X) - 1], X[len(X) - 1:]

    model = AutoReg(train, lags=29)

    model_fit = model.fit()

    predictions = model_fit.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

    vara = predictions.to_json()

    vara = json.loads(vara)

    valueOfPrediction = vara["999"]

    p = {"newDatas": valueOfPrediction}

    UrlToPostTemp = "http://localhost:49743/api/Result/HumidityResult"

    headers = {'content-type': 'application/json'}

    response = requests.post(UrlToPostTemp, headers=headers, data=json.dumps(p), verify=False)

    print(response, "Humidity")


def predictCO2():
    UrltoGetTem = "http://localhost:49743/api/Data/CO2Data"

    weather = pd.read_json(UrltoGetTem)

    X = weather.dataValue

    train, test = X[1:len(X) - 1], X[len(X) - 1:]

    model = AutoReg(train, lags=29)

    model_fit = model.fit()

    predictions = model_fit.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

    vara = predictions.to_json()

    vara = json.loads(vara)

    valueOfPrediction = vara["499"]

    p = {"newDatas": valueOfPrediction}

    UrlToPostTemp = "http://localhost:49743/api/Result/CO2Result"

    headers = {'content-type': 'application/json'}

    response = requests.post(UrlToPostTemp, headers=headers, data=json.dumps(p), verify=False)

    print(response, "CO2")


def predictAirQuality():
    UrltoGetTem = "http://localhost:49743/api/Data/AirQualityData"

    weather = pd.read_json(UrltoGetTem)

    X = weather.dataValue

    train, test = X[1:len(X) - 1], X[len(X) - 1:]

    model = AutoReg(train, lags=29)

    model_fit = model.fit()

    predictions = model_fit.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

    vara = predictions.to_json()

    vara = json.loads(vara)

    valueOfPrediction = vara["999"]

    p = {"newDatas": valueOfPrediction}

    UrlToPostTemp = "http://localhost:49743/api/Result/AirQualityResult"

    headers = {'content-type': 'application/json'}

    response = requests.post(UrlToPostTemp, headers=headers, data=json.dumps(p), verify=False)

    print(response, "AirQuality")


def predictBurningGas():
    UrltoGetTem = "http://localhost:49743/api/Data/BurningGasData"

    weather = pd.read_json(UrltoGetTem)

    X = weather.dataValue

    train, test = X[1:len(X) - 1], X[len(X) - 1:]

    model = AutoReg(train, lags=29)

    model_fit = model.fit()

    predictions = model_fit.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

    vara = predictions.to_json()

    vara = json.loads(vara)

    valueOfPrediction = vara["499"]

    p = {"newDatas": valueOfPrediction}

    UrlToPostTemp = "http://localhost:49743/api/Result/BurningGasResult"

    headers = {'content-type': 'application/json'}

    response = requests.post(UrlToPostTemp, headers=headers, data=json.dumps(p), verify=False)

    print(response, "BurningGas")


schedule.every(10).minutes.do(predictTempature)
schedule.every(10).minutes.do(predictHumidity)
schedule.every(10).minutes.do(predictCO2)
schedule.every(10).minutes.do(predictAirQuality)
schedule.every(10).minutes.do(predictBurningGas)

while True:
    schedule.run_pending()
    time.sleep(1)
```
